```python
import pandas as pd

df = pd.read_csv("zomato.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['rate (out of 5)', 'avg cost (two people)'])

df.isnull().sum()

# I've decided to drop the rows which contained null values for rating and avg cost as the number of null values are 125,
# which is insignificant in comparison to our dataset. Also either changing the rating to the average or '0' would skew the results.
# Similarly with avg cost there is no way of estimating an avg cost for a restaurant as they may vary vastly.
```

```python
df = df.drop_duplicates()
```

```python
df['restaurant name'] = df['restaurant name'].str.strip()
df['area'] = df['area'].str.title()

# convert avg cost into int for correct calculations
df['avg cost (two people)'] = df['avg cost (two people)'].astype(int)
```

```python
df['cost_per_person'] = df['avg cost (two people)']/2

def categorize_rating(rating):
  if rating >= 4.5:
    return "Very Good"
  elif rating >= 3.5:
    return "Good"
  elif rating >= 2.5:
    return "Average"
  else:
    return "Poor"

df['rating_category'] = df['rate (out of 5)'].apply(categorize_rating)
```

```python
df.to_csv("cleaned_dataset.csv")
```
